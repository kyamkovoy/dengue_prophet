#### Define some functions to make everything easier:

```python
# loads the error dataframe
def load_error_df(folder):
    # load the data
    file_path = folder + '/monthly_errors.csv'
    error_df = pd.read_csv(file_path)
    error_df['province'] = error_df['province'].astype(int)
    error_df['year'] = error_df['year'].astype(int)
    error_df['peak_month_error'] = error_df['peak_month_error'].astype(int)
    error_df = error_df.loc[error_df['year'] != 2007]
    
    return error_df
```

```python
# makes a graph for specified province
def make_a_graph(folder, province):
    error_df = load_error_df(folder)
    prov_df = error_df.loc[error_df['province'] == province]
    ax = sns.barplot(x='year', y='monthly_case_rmse', data=prov_df, palette="Blues_d", ci=0)
    ax.set(xlabel='Year', ylabel='RMSE for Monthly Case Counts')
    plt.title(province_names[province])
    plt.show()
```

```python
# just load all of the error dfs for quick access

hist_avg_folder = '../../output/hist_avg'
jan_default_prophet_folder = '../../output/jan_time0/default_prophet'
jan_cp_prophet_folder = '../../output/jan_time0/cp_prophet'
jul_default_prophet_folder = '../../output/jul_time0/default_prophet'
jul_cp_prophet_folder = '../../output/jul_time0/cp_prophet'

avg_error = load_error_df(hist_avg_folder)
jan_dp_error = load_error_df(jan_default_prophet_folder)
jan_cpp_error = load_error_df(jan_cp_prophet_folder)
jul_dp_error = load_error_df(jul_default_prophet_folder)
jul_cpp_error = load_error_df(jul_cp_prophet_folder)
```

### Compare performance of the models numerically

```python
# this function gets all of the AVERAGE errors for the specified target for all provinces

def avg_errors(province, which_error, which_time0, output = False):
    
    if which_time0 == 'jan':
        dp_error = jan_dp_error.loc[jan_dp_error['province'] == province]
        cpp_error = jan_cpp_error.loc[jan_cpp_error['province'] == province]
        prov_avg_error = avg_error.loc[avg_error['province'] == province]
        
        mpd_avg = dp_error[which_error].mean()
        mpcp_avg = cpp_error[which_error].mean()
        mavg_avg = prov_avg_error[which_error].mean()
    
    if which_time0 == 'jul':
        dp_error = jul_dp_error.loc[jul_dp_error['province'] == province]
        cpp_error = jul_cpp_error.loc[jul_cpp_error['province'] == province]
        prov_avg_error = avg_error.loc[avg_error['province'] == province]
        
        mpd_avg = dp_error[which_error].mean()
        mpcp_avg = cpp_error[which_error].mean()
        mavg_avg = prov_avg_error[which_error].mean()

    if output == True:
        print('default prophet errors for ' + which_error)
        print('   ' + str(mpd_avg))
        
        print()
        print('prophet w/ changepoint prior = 0.5 errors for ' + which_error)
        print('   ' + str(mpcp_avg))
        
        print()
        print('historical average errors for ' + which_error)
        print('   ' + str(mavg_avg))
    
    return [mpd_avg, mpcp_avg, mavg_avg]
```

```python
# use this one for graphing ALL PROVINCES

def avg_errors_province_ys(which_error, which_time0, output = False):
    
    single_prov_errors = []
    for province in provinces:
        if which_time0 == 'jan':
            dp_error = jan_dp_error.loc[jan_dp_error['province'] == province]
            cpp_error = jan_cpp_error.loc[jan_cpp_error['province'] == province]
            prov_avg_error = avg_error.loc[avg_error['province'] == province]

            mpd_avg = dp_error[which_error].mean()
            mpcp_avg = cpp_error[which_error].mean()
            mavg_avg = prov_avg_error[which_error].mean()

        if which_time0 == 'jul':
            dp_error = jul_dp_error.loc[jul_dp_error['province'] == province]
            cpp_error = jul_cpp_error.loc[jul_cpp_error['province'] == province]
            prov_avg_error = avg_error.loc[avg_error['province'] == province]

            mpd_avg = dp_error[which_error].mean()
            mpcp_avg = cpp_error[which_error].mean()
            mavg_avg = prov_avg_error[which_error].mean()

        if output == True:
            print('default prophet errors for ' + which_error)
            print('   ' + str(mpd_avg))

            print()
            print('prophet w/ changepoint prior = 0.5 errors for ' + which_error)
            print('   ' + str(mpcp_avg))

            print()
            print('historical average errors for ' + which_error)
            print('   ' + str(mavg_avg))
    
        single_prov_errors.append([mpd_avg, mpcp_avg, mavg_avg])
    
    errors = pd.DataFrame(single_prov_errors, columns=['dp', 'cpp', 'hist_avg'])
    return errors
```
